```python
x_train = torch.FloatTensor([[1],[2],[3]]).cuda()
y_train = torch.FloatTensor([[2],[4],[6]]).cuda()

W = torch.zeros(1, requires_grad=True).cuda() # requires_grad를 True로 설정해 학습할 것이라 명시

# y = Wx
# W : weight

lr = 0.01

epochs = 100
for epoch in range(1, epochs + 1):

  hypothesis = x_train * W
  
  cost = torch.mean((hypothesis - y_train)**2) # cost(MSE)  계산
  gradient = 2*torch.mean((W * x_train - y_train) * x_train) # cost에 대한 w미분(Loss에 대해 w가 미치는 영향)

  W = W - lr*gradient # cost gradient로 W 갱신

  if epoch%10 == 0:
    print(f"{epoch} : Loss = {cost}")

print("Train End")

inference_data = torch.FloatTensor([1.5, 9.4, 10.4, 1, 2, 3]).cuda()

result = inference_data * W

print(W)
print(result)
```
