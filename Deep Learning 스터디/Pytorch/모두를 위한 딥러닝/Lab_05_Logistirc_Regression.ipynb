```python
x_data = [[1,2],[2,3],[3,1],[4,3],[5,3],[6,2]]
y_data = [[0],[0],[0],[1],[1],[1]]

x_train = torch.FloatTensor(x_data)
y_train = torch.FloatTensor(y_data)

print(x_train.shape)
print(y_train.shape)
```

```python
w = torch.zeros((2,1), requires_grad=True)
b = torch.zeros(1, requires_grad=True)

# hypothesis = 1 / (1+ torch.exp(-(x_train.matmul(w) + b)))
hypothesis = torch.sigmoid(x_train.matmul(w) + b)

print(hypothesis)
print(hypothesis.shape)
```

```python
# losses = -(y_train * torch.log(hypothesis) + (1-y_train)*torch.log(1 - hypothesis))
# cost = losses.mean()
cost = F.binary_cross_entropy(hypothesis, y_train)

print(cost)
```

```python
w = torch.zeros((2,1), requires_grad=True)
b = torch.zeros(1, requires_grad=True)

optimizer = optim.SGD([w,b], lr= 1)

for epoch in range(1, 1001):

  hypothesis = torch.sigmoid(x_train.matmul(w) + b) # 활성화 함수 통과 = 모델의 output
  cost = F.binary_cross_entropy(hypothesis, y_train) # Loss 함수

  optimizer.zero_grad()
  cost.backward()
  optimizer.step()

  if epoch % 100 == 0:
    print(f"Epoch :{epoch}/1000, Loss : {cost.item()}")
```

```python
class BinaryClassifier(nn.Module):
  def __init__(self):
    super().__init__()
    self.linear = nn.Linear(2,1) # M x 8 차원의 입렬과 1개의 Bias
    self.sigmoid = nn.Sigmoid()

  def forward(self, x):
    return self.sigmoid(self.linear(x))


def calculate_acc(output, label):
  predict = output >= torch.FloatTensor([0.5])
  predict = predict.type(torch.IntTensor)

  acc = (predict == label).type(torch.IntTensor)
  acc = torch.sum(acc)

  return acc

x_data = [[1,2],[2,3],[3,1],[4,3],[5,3],[6,2]]
y_data = [[0],[0],[0],[1],[1],[1]]

x_train = torch.FloatTensor(x_data)
y_train = torch.FloatTensor(y_data)


model = BinaryClassifier()
optimizer = optim.SGD(model.parameters(), lr = 1)

total_epoch = 100

for epoch in range(1, total_epoch+1):

  output = model(x_train)
  loss = F.binary_cross_entropy(output, y_train)

  optimizer.zero_grad()
  loss.backward()
  optimizer.step()

  if epoch % 10 == 0:
    print(f"Epoch : {epoch}/{total_epoch}, Loss : {loss.item()}, Acc : {calculate_acc(output, y_train) / len(y_train) * 100} ")

```
