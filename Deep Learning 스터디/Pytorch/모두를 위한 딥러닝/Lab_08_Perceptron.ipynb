```python
x_data = [[0,0],
          [0,1],
          [1,0],
          [1,1]]
y_data = [[0],[1],[1],[0]]

x_train = torch.Tensor(x_data)
y_train = torch.Tensor(y_data)

class Classifier(nn.Module):
  def __init__(self):
    super().__init__()
    self.layer = nn.Sequential(
        nn.Linear(2, 10),
        nn.Sigmoid(),
        nn.Linear(10, 1),
        nn.Sigmoid()
    )

  def forward(self, x):
    
    return self.layer(x)

model = Classifier()

optimzier = optim.SGD(model.parameters(), lr= 1)
loss_function = torch.nn.BCELoss()

for epoch in range(1, 10000):

  predict = model(x_train)
  loss = loss_function(predict, y_train)

  optimzier.zero_grad()
  loss.backward()
  optimzier.step()
  if epoch% 1000 == 0:
    print(epoch ,loss.item())

print(model(x_train))
```
