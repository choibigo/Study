```python
t1 = torch.FloatTensor([[3,3]])
t2 = torch.FloatTensor([[2,2]])

print(t1 + t2)

print(t1 + torch.Tensor([100])) # 100 -> [[100, 100]]

m1 = torch.Tensor([[1,2]])     # [[1,2]] -> [[1,2], [1,2]] 
m2 = torch.Tensor([[3], [4]])  # [[3],[4]] -> [[3,3],[4,4]] # 두 행렬을 더큰 차원으로 바꿔서 연산을 수행한다.

print(m1 + m2)
```

- 일반적인 곱(*)을 사용하면 자동으로 차원을 변경하여 각각 원소별 곱을 수행한다.
- Multiplication은 일반적인 행렬의 곱 연산을 수행한다.

```python
m1 = torch.Tensor([[1,2],[3,4]])     
m2 = torch.Tensor([[1], [2]])
print(m1.matmul(m2))

print(m1*m2)
```

- mean() 함수는 LongTensor에 대해서 수행하지 못한다.

```python
t1 = torch.Tensor([[1,2],[3,4]])
t2 = torch.LongTensor([1,2])

print(t1.mean()) # 전체 원소에 대해 평균 수행
# print(t2.mean())

print("====== Mean ======")
print(t1.mean(dim = 0)) # row에 대해 평균 수행
print(t1.mean(dim = 1)) # col에 대해 평균 수행

print("====== Sum ======")
print(t1.sum())
print(t1.sum(dim = 0))
print(t1.sum(dim = 1))

print("====== Max ======")
print(t1.max())
print(t1.max(dim = 0))
print(t1.max(dim = 1))
```

- View 는 numpy의 Reshape 함수 이다.

```python
t = np.array([[[0,1,2],
               [3,4,5]],
               [[10,11,12],
               [13,14,15]]])

ft = torch.Tensor(t)
print(ft.shape)

print(ft.view(-1, 3).shape)
print(ft.view(2, 1, -1).shape) # view 내부 원소 합이 기존 데이터 shape의 곱과 같으면 된다.
```

- Squeeze 를 통해 dim 이 1인 차원을 없앨 수 있다.
- Unsqueeze를 통해 dim 이 1인 차원을 추가 한다.

```python
t = np.array([[0],[1],[2]])
ft = torch.Tensor(t)

print(ft.shape)
print(ft.squeeze().shape) # 1인 차원을 없애 준다.
print(ft.unsqueeze(dim=0).shape) # 1인 차원을 추가 한다.
```

Type Casting

```python
lt = torch.LongTensor([1,2,3,4,])
print(lt)
print(lt.float())
bt = (lt == 3)
print(bt) # 조건 문을 토앻 byteTensor를 생성할 수 있다.
```

- Concatenate 두 텐서 합치기

```python
x = torch.Tensor([[1,2],[3,4]])
y = torch.Tensor([[11,12],[13,14]])

print(torch.cat([x,y], dim = 0)) # row 쪽으로 추가
print(torch.cat([x,y], dim = 1)) # col 쪽으로 추가
```

Stacking

```python
x = torch.Tensor([[1,2],[3,4]])
y = torch.Tensor([[11,12],[13,14]])
z = torch.Tensor([[111,12],[113,114]])

print(torch.stack([x,y,z]))
print(torch.stack([x,y,z]).shape) # 차원이 새로 추가 된다. 리스트를 만든것
```
