# 텐서 초기화와 데이터 타입


초기화 되지 않는 텐서

```python
x = torch.empty(4,2) # (row, col)
print(x)
```

무작위로 초기화된 텐서

```python
x = torch.rand(4,2)
print(x)
```

데이터 타입이 long이고, 0으로 채워진 텐서

```python
x = torch.zeros(4,2, dtype=torch.long)
print(x)
```

사용자가 입력한 값으로 텐서 초기화

```python
x = torch.tensor([3, 33,33])
print(x)
```

2 x 4크기, double type 1로 채워진 텐서


```python
x = x.new_ones(2, 4, dtype=torch.double)
print(x)
```

x와 같은 크기, float 타입, 무작위로 채워진 텐서
- rand_like 사용

```python
x = torch.rand_like(x, dtype=torch.float)
print(x)
print(x.size())
```

Data Type

```python
ft = torch.FloatTensor([1,2,3])
print(ft, ft.dtype)

print(ft.short())
print(ft.int())
print(ft.long())

it = torch.IntTensor([1,2,3])
print(it, it.dtype)
print(it.float())
print(it.double())
```

CUDA Tensors
- .to 메소드를 사용하여 텐서를 어떠한 장치(cpu, gpu)로도 옮길 수 있음

```python
x = torch.rand([2,4])
print(x, x.dtype)

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(device)

y = torch.ones_like(x, device=device)
print(y, y.dtype)

x = x.to(device)
print(x)

z = x + y
print(z)

print(z.to('cpu', torch.double))
```

0D Tensor(Scalar)
- 하나의 숫자를 담고 있는 텐서
- 축과 형상이 없음

```python
t0 = torch.tensor(0)
print(t0.ndim)
print(t0.shape)
print(t0)
```

1D Tensor(Vector)

- 값들을 저장한 리스트와 유사한 텐서
- 하나의 축 존재

```python
t1 = torch.tensor([1,2,3])
print(t1.ndim)
print(t1.shape)
print(t1)
```

2D Tensor(Matrix)

- 행렬과 같은 모양으로 두개의 축이 존재
- 일반적으인 수치, 통계 데이터 셋이 해당
- 주로 Sample과 Features를 가진 구조로 사용

```python
t2 = torch.tensor([[1,2,3], [4,5,6]])
print(t2.ndim)
print(t2.shape)
print(t2)
```

# 텐서의 조작

인덱싱 : Numpy 처럼 인덱싱 형태로 사용가능

```python
x = torch.Tensor([[1,2],
                 [3,4]])

print(x)
print(x[0][0])
print(x[0, 0])
print(x[1, :])
```

view : 텐서의 크기나 모양을 변경
- 기본적으로 변경 전 후에 텐서안의 원소개수가 유지 되야함
- -1로 설정되면 계산을 통해 해당 크기 값을 유추

```python
x = torch.rand(4,5)
print(x)
y = x.view(20)
print(y)
z = x.view(5, -1)
print(z)
```

item : 텐서에 값이 하나라도 존재하면 굿자 값을 얻을 수 있다
- 스칼라값 하나만 있을떄만 사용할 수 있다.

```python
x = torch.rand(1)
print(x)
print(x.item())
print(x.dtype)
```

squeeze : 차원을 축소(제거)
- 맨앞이 1차원이라면 차원을 하나 없애도 같은 의미이다.
- 이떄 squeeze를 사용

```python
tensor = torch.rand(1,3,3)
print(tensor)
print(tensor.shape)

tensor2 = tensor.squeeze()
print(tensor2)
print(tensor2.shape)
```

unsqueeze : 차원을 증가(생성)

```python
tensor = torch.rand(3,3)
print(tensor)
print(tensor.shape)

tensor2 = tensor.unsqueeze(1) # 2번쨰 차원에 대해 1차원을 늘려라. 
print(tensor2)
print(tensor2.shape)
```

stack : 텐서간 결합

```python
x = torch.FloatTensor([1,4])
y = torch.FloatTensor([2,5])
z = torch.FloatTensor([3,6])

print(torch.stack([x,y,z]))
```

cat : 텐서를 결합하는 메소드(concatenate)
- stack과 유사하지만 쌓을 dim이 존재
- 해당 차원을 늘려준 후 결합

```python
a = torch.rand(1,3,3)
b = torch.rand(1,3,3)
print(a, a.shape)
print(b, b.shape)
# c = torch.cat((a,b), dim = 0) # 1번쨰 차원을 기준으로 합친다.
# c = torch.cat((a,b), dim = 1) # 2번쨰 차원을 기준으로 합친다.
c = torch.cat((a,b), dim = 2) # 3번쨰 차원을 기준으로 합친다.
print(c, c.shape)
```

chunck : 텐서를 여러개로 나눌 떄 사용(몇개로 나눌 것인가?)

```python
tensor = torch.rand(3,6)
print(tensor)
t1, t2, t3 = torch.chunk(tensor, 3, dim=1) # 3개로 2번쨰 차원을 기준으로 나눠줘
print(t1)
print(t2)
print(t3)
```

split : chunck와 동일하난 기즌이 다름(텐서의 크기는 몇으로 나눌꺼냐?)

```python
tensor = torch.rand(3, 6)
t1, t2 = torch.split(tensor, 3, dim = 1)
print(t1)
print(t2)
```

# Torch <==> Numpy

Torch Tensor를 Numpy array로 변환 가능
- numpy()
- from_numpy()

Tensor가 CPU상에 있다면 Numpy 배열은 메모리 공간을 공유 함으로 하나가 변하면 다른 하나도 변함

```python
a = torch.ones(2,4)
print(a)

b = a.numpy()
print(b)

a.add_(3)

print(a)
print(b)
```

```python
a = np.ones(7)
b = torch.from_numpy(a)

np.add(a, 1, out = a)

print(a)
print(b)
```
